```python
import pandas as pd
import numpy as np
import os

img_size = 224
num_classes = 16   # we will select top 16 most fequently occuring breeds to save computatinal time
seed = 1987

data_dir = 'C:/Users/Sakshee/Documents/DATASETS/Dog_breed_identification/train'
test_dir = 'C:/Users/Sakshee/Documents/DATASETS/Dog_breed_identification/test'

labels = pd.read_csv(os.path.join(data_dir,'C:/Users/Sakshee/Documents/DATASETS/Dog_breed_identification/labels.csv'))
submission = pd.read_csv(os.path.join(test_dir,'C:/Users/Sakshee/Documents/DATASETS/Dog_breed_identification/sample_submission.csv'))
#Maps the id in the csv file to its corresponding image id in the data_dir folder for both csvs
labels.head()
```

HOW ALL THE FUNCTIONS USED WORK

```python
#it groups all the other features of the data frame in accordance with the feature to group by
S_B_L = labels.groupby('breed')
#to show the first members(row of id) of each group of breeds
S_B_L.first()
```

```python
#.count().sort_values() it sorts the breeds with highest number of ids to least(since ascending = False)

selected_breed_dataframe = labels.groupby('breed').count().sort_values(by='id', ascending=False).head(num_classes)
selected_breed_dataframe.head()
```

```python
#by adding .index... the variable becomes a index object and returns row labels(in this case the breeds) which me then print out as a list

selected_breed_list = list(labels.groupby('breed').count().sort_values(by='id', ascending=False).head(num_classes).index)
selected_breed_list
```

```python
labels = labels[labels['breed'].isin(selected_breed_list)]
labels['target'] = 1                                          #all breeds that are top 16 are targetted as 1
labels.head()
```

```python
group = labels.groupby(by='breed', as_index=False).agg({'id': pd.Series.nunique})
group

#nunique returns the number of unique elements in the object eg {1,2,3,7,7} then unique elements is 4
#so in this example we are basically counting how many unique ids are present for a certain breed
```

```python
#.pivot() reshapes the dataframe.
# the first parameter=index: sets the row labels (ie id)
# the second parameter=columns: are the column labels (ie breed)
#the third parameter=values: are used to populate the new dataframe (ie target)
labels_pivot = labels.pivot('id', 'breed', 'target').reset_index().fillna(0)
labels_pivot.head()
```

```python
np.random.seed(seed=seed)                           #same random numbers will generate each time
rnd = np.random.random(len(labels))
rnd
```

```python
train_idx = rnd < 0.8         #returns an array of True values when condition is satisfied and false when the condition is not
valid_idx = rnd >= 0.8        #vice-versa
print(train_idx)
print(train_idx.shape)
```

```python
#selected_breed_list = 'breed1', 'breed2'......'breed16'
y_train = labels_pivot[selected_breed_list].values
print(y_train)
print(y_train.shape)
```

```python
ytr = y_train[train_idx]
yv = y_train[valid_idx]
print(ytr)
print(ytr.shape)
print(yv.shape)
#we seperated the training data into two arrays correspoding to their train_idx being true or false
#1409 training samples
#368 validation samples.....total 1777 samples

#ytr and yv are the output labels. For eg. In cats vs dogs classification (20000,1) was ytr and (5000,1) was yv
```

```python
#now to load the desired images of the selected 16 breeds as a 4-D day array for input
#xtr = (1409,224,224,1)
#xv = (368,224,224,1)
from keras.preprocessing.image import img_to_array, load_img
import cv2

img_id = labels_pivot['id'].values

X = []
for ID in img_id:
    img = load_img(os.path.join(data_dir,'%s.jpg' %ID), grayscale = True, target_size=(img_size,img_size))
    image = (img_to_array(img))
    X.append(image)
    
X = np.array(X)
X.shape
```

```python
xtr = X[train_idx]
xv = X[valid_idx]
print(xtr.shape)
print(xv.shape)
```

```python
xtr = xtr.astype('float32')
xv = xv.astype('float32')

xtr /=255
xv /=255 
```

```python
import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense

def baseline_model():
    model = Sequential()
    
    model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 1), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    
    model.add(Dropout(0.2))
    
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

model = baseline_model()
model.compile(optimizer='RMSprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(xtr, ytr, validation_data=(xv,yv), epochs=5, verbose=1)
```

```python
test_dir = 'C:/Users/Sakshee/Documents/DATASETS/Dog_breed_identification/test'

test = []
for img_id in os.listdir(test_dir):
    img = load_img(os.path.join(test_dir, img_id), grayscale = True, target_size=(img_size,img_size))
    image = (img_to_array(img))
    test.append(image)
    
test = np.array(test)
test.shape
```

```python
predictions = model.predict(test)

predictions.shape
```
